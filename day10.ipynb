```python
CHUNK_SCORE = {
    ")": 3,
    "]": 57,
    "}": 1197,
    ">": 25137,
}
CHUNK_RULES = {
    "[": "]",
    "(": ")",
    "<": ">",
    "{": "}"
}
CHUNK_CLOSING_SCORE = {    
    ")": 1,
    "]": 2,
    "}": 3,
    ">": 4,
}
```

```python
def first_incorrect_char(line: str) -> str:
    open_chunks = []
    for char in line:
        if char in CHUNK_RULES:
            open_chunks.append(char)
        elif CHUNK_RULES[open_chunks[-1]] == char:
            open_chunks.pop()
        else:
            return char
        
    return ""
        
```

```python
incorrect_chars = []
for line in demo_input:
    incorrect_char = first_incorrect_char(line)
    if incorrect_char:
        incorrect_chars.append(incorrect_char)
```

```python
#demo answer
sum(CHUNK_SCORE[c] for c in incorrect_chars)
```

```python
def complete_closing_seq(line: str) -> str:
    open_chunks = []
    for char in line:
        if char in CHUNK_RULES:
            open_chunks.append(char)
        elif CHUNK_RULES[open_chunks[-1]] == char:
            open_chunks.pop()
        else:
            return char
        
    # what is left in open_chunks is unfinished
    return [CHUNK_RULES[c] for c in open_chunks[::-1]]
        
```

```python
def score_closing_seq(seq: list[str]) -> int:
    score = 0
    for c in seq:
        score *= 5
        score += CHUNK_CLOSING_SCORE[c]
    return score
```

```python
score_closing_seq(complete_closing_seq(incomplete_demo_input[0]))
```

```python
scores = []
for line in incomplete_demo_input:
    score = score_closing_seq(complete_closing_seq(line))
    scores.append(score)

sorted(scores)[len(scores) // 2]
```
