```python
boards = []
board = []
for line in boards_as_str.splitlines():
    if line == '':
        boards.append(np.array(board, dtype=int))
        board = []
        continue
    
    board.append(list(map(int, line.split())))
```

```python
boards
```

```python
drawn_numbers = numbers[:5]
winner_board = None
for number in numbers[5:]:
    for board in boards:
        for i in range(5):
            if set(drawn_numbers).issuperset(set(board[i,:])) or set(drawn_numbers).issuperset(set(board[:,i])):
                print("found a winner!")
                winner_board = board
                break
    
    if winner_board is not None:
        break
    
    drawn_numbers.append(number)
```

```python
remaining_vals = [x for x in winner_board.flatten() if x not in drawn_numbers]
```
